```python
import numpy as np
import matplotlib.pyplot as plt

# Read processing times from file
with open('processing_times.txt', 'r') as file:
    processing_times = [int(line.strip()) for line in file]

# Convert the list to a NumPy array
processing_times = np.array(processing_times)

# Calculate the CDF
sorted_times = np.sort(processing_times)
cdf = np.arange(1, len(sorted_times) + 1) / len(sorted_times)

# Plot the CDF
plt.plot(sorted_times, cdf * 100)  # Multiply by 100 to convert to percentage
plt.xlabel('Processing Time (ms)')
plt.ylabel('Percentage')
plt.title('CDF of Processing Times')
plt.grid(True)
plt.show()
```
